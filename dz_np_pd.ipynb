Задача 1. Создать матрицу размером 8х8 элементов, состоящую из нулей.Заполнить эту матрицу значениями, расположенными в шахматном порядке.Задачу решить через срезы массива NumPy. Вывести полученную матрицу на экран и в файл.

```python
matrix = np.zeros((8,8), dtype=int)
print(matrix)
```

```python
matrix[1::2,::2] = 1
matrix[::2,1::2] = 1
print(matrix)
np.savetxt("array.csv", matrix, delimiter=",")
```

Задача 2. Средствами NumPy сформируйте вектор из 2000 случайных величин,распределенных по нормальному закону с математическим ожиданием 0 и дисперсией 5. Средствами NumPy вычислите для полученного вектора экспериментальное значение математического ожидания(среднее значение) и дисперсию.


```python
normal_array = np.random.normal(loc=0, scale=np.sqrt(5), size=2000)
print(normal_array)
print(f'Эксперементальное значение математического ожидания {normal_array.mean()}')
print(f'Эксперементальное значение дисперсии {normal_array.var()}')
```
